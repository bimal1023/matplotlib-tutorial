## Categorical Plotting (Bar, Scatter, Line)

This example demonstrates three different ways to visualize categorical data using Matplotlib.


```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

## Damped Oscillation Example

This example plots a damped cosine wave and compares it with a regular cosine wave.


```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

## Training vs Validation Loss Curve

This example visualizes model performance over epochs using training and validation loss.


```python
train_losses = [
    1.95, 1.42, 1.10, 0.88, 0.72,
    0.60, 0.52, 0.45, 0.39, 0.34,
    0.30, 0.27, 0.24, 0.22, 0.20
]

val_losses = [
    1.80, 1.35, 1.05, 0.90, 0.78,
    0.68, 0.60, 0.55, 0.50, 0.47,
    0.45, 0.44, 0.43, 0.43, 0.44
]
```

## Confusion Matrix (Heatmap Visualization)

This example visualizes model classification results using a confusion matrix.


```python
import seaborn as sns

conf_matrix = np.array([
    [50, 2, 1],
    [3, 45, 5],
    [0, 4, 48]
])

plt.figure(figsize=(6,5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

## ROC Curve & AUC

This example evaluates a binary classifier using the ROC curve and AUC score.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# True labels
y_true = np.array([0, 0, 1, 1, 0, 1, 0, 1, 1, 0])

# Predicted probabilities (model outputs)
y_scores = np.array([0.1, 0.4, 0.8, 0.9, 0.3, 0.75, 0.2, 0.85, 0.95, 0.05])
```

```python
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)
```

```python
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0,1], [0,1], linestyle="--")  # random classifier line

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.show()
```

## PrecisionRecall Curve

This example evaluates a binary classifier using the PrecisionRecall curve.


```python
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
```

```python
plt.figure(figsize=(6,5))
plt.plot(recall, precision)

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("PrecisionRecall Curve")
plt.grid(True)
plt.show()
```

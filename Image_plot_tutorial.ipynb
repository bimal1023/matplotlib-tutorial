## Visualizing the Image using Matplotlib

Importing the necessary Libraries

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
%matplotlib inline
```

Using pillow library to load image and converting it into Numpy Arrays

```python
img=np.asarray(Image.open("/Users/bimalkumal/Downloads/flower/train/dandelion/10683189_bd6e371b97_jpg.rf.213ffbff3870cabf5f0701a25c8a1b57.jpg"))
print(img)
```

Checking the shape of img

```python
print(img.shape)
```

Plotting numpy arrays as images


```python
imgplot=plt.imshow(img)
```

Applying Pseudocolor Schemes to image plots

* Pseudo is only relevant to single-channel, grayscale, luminosity images

```python
lum_img=img[:,:,0]
plt.imshow(lum_img)
```

```python
plt.imshow(lum_img,cmap="hot")
```

```python
imgplot=plt.imshow(lum_img)
imgplot.set_cmap("nipy_spectral")
```

* Adding color bar to the figure

```python
imgplot=plt.imshow(lum_img)
plt.colorbar()
```
